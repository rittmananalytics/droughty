```python
import markdown
import os

try:
    with open("/Users/lewischarlesbaker/Documents/GitHub/droughty/warehouse_docs/field_descriptions.md") as f:
        htmlmarkdown=markdown.markdown( f.read() )
        
    def get_next_words(text, match, sep=' '):
        words = iter(text.split(sep))
        for word in words:
            if word == match:
                yield next(words)
                
    text = htmlmarkdown
    words = get_next_words(text, 'docs')

    described_columns_list = []

    for w in words:
        
        described_columns_list.append (w)
except:
    described_columns_list = []
```
